```python
import pandas as pd
```

**1) Reading data extracted from csv downloaded from Amazon.com** 

 *Data contains all orders placed on Amazon by Customer from year 2017*

```python
sidd_orders = pd.read_csv('amazon-orders.csv')
```

```python
sidd_orders.head(3)
```

---
**2) Basic data cleansing to remove nulls and change datatype of column**

```python
df = sidd_orders.fillna(0)
```

```python
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','', regex =True).astype(float)
```

```python
df.head()
```

---
**3) Some high-level insights into the data**

```python
total_spent = print("You have spent a total of " + str(sum(df["Total Charged"])) + " Dollars on Amazon Shopping since 2017")
total_spent
```

```python
average_spent = print("You have spent an average of " + str(round(df["Total Charged"].mean(),2)) + " Dollars on Amazon Shopping since 2017")
average_spent
```

```python
most_spent = print("Your most expensive order is " + str(max(df["Total Charged"])) + " Dollars on Amazon Shopping since 2017")
most_spent
```

---

**4) Visualizing the data using Matplotlib** 

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.head()
```

```python
%matplotlib inline
```

```python
df.plot.bar(x='Order Date',y='Total Charged', rot=90, figsize=(20,10))
```

```python
orders_daily = df.groupby('Order Date').sum()["Total Charged"]
orders_daily.head()
```

```python
orders_daily.plot.bar(figsize = (20,10), title = 'Daily spending on Amazon', ylabel='Dollars $', Color = 'red')
```
